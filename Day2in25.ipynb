### Handling Missing Values

To add some complexity to the book's examples I used houses prices dataset.

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_boston
data = load_boston()
```

```python
data.data
```

Since I am interested in numerical values to identify the missing values, no need to include the house price (target of the dataset), working will all or some of the features does the job.

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df
```

```python
df.isnull()
```

```python
# get the column name that has null values
df.isnull().any()
```

```python
# drop some of the values to apply the following steps

df['CRIM'] = df['CRIM'].drop(labels=[1,3,20], axis=0)
df['AGE'] = df['AGE'].drop(labels=[100,300,500], axis=0)
df['PTRATIO'] = df['PTRATIO'].drop(labels=[63,280,462], axis=0)
df['B'] = df['B'].drop(labels=[440,441,442], axis=0)
```

### Identifying Missing values

```python
# print the rows+columns that contains nan values
df[df.isnull().any(axis=1)]
```

### Data Infilling

```python
# filling nan with zeros
df_0 = df.fillna(0)
df_0[df_0.isnull().any(axis=1)]
```

```python
df[438:445]
```

```python
# filling nan with the previous data row
df_pad = df.fillna(method='pad')
df_pad[438:445]
```

```python
# filling nan with the previous data row
df_bfill = df.fillna(method='bfill')
df_bfill[438:445]
```

### Droping missing values

```python
df_drop = df.dropna()
df_drop[df_drop.isnull().any(axis=1)]
```

### Sorting iris dataset by petal length

```python
iris.rename(columns={'target':'Species'}, inplace=True)
```

```python
iris['Species'] == 1
```

```python
def assign_names(item):
  if item == 0:
    return 'setosa'
  elif item == 1:
    return 'versicolor'
  else:
    return 'virginica'
```

```python
iris['target'] = iris['Species'].apply(assign_names)
```
